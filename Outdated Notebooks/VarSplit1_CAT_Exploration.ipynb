```python
import seaborn as sns
from matplotlib import pyplot as plt
import plotly.express as px
```

```python
def histograms_plot(dataframe, features, rows, cols, binNum, xsize=20, ysize=20):

    fig=plt.figure(figsize=(xsize,ysize))
    for i, feature in enumerate(features):
        ax=fig.add_subplot(rows,cols,i+1)
        dataframe[feature].hist(bins=binNum,
                                ax=ax,
                                facecolor='skyblue')
        ax.set_title(feature,color='purple')

    fig.tight_layout()  
    plt.show()
```

```python
def violins_plot(dataframe, features, yvar, rows, cols,xsize=20, ysize=20):
    auxdf=dataframe.copy()
    auxdf["dummyVar"]=0
    fig=plt.figure(figsize=(xsize,ysize))
    for i, feature in enumerate(features):
        ax=fig.add_subplot(rows,cols,i+1)
        sns.violinplot(ax=ax,x=auxdf["dummyVar"], y=yvar, hue=dataframe[feature],split=True, alpha=0.7)
        ax.set(xlabel=None)
        ax.set(xticklabels=[])
        
    fig.tight_layout()  
    plt.show()
```

Except for itt_all_i, the rest of CAT variables can be dummy encoded, since they are binary propositions. 

From https://machinelearningmastery.com/one-hot-encoding-for-categorical-data/:
In addition to being slightly less redundant, a dummy variable representation is required for some models.

For example, in the case of a linear regression model (and other regression models that have a bias term), a one hot encoding will case the matrix of input data to become singular, meaning it cannot be inverted and the linear regression coefficients cannot be calculated using linear algebra. For these types of models a dummy variable encoding must be used instead.

"If the model includes an intercept and contains dummy variables […], then the […] columns would add up (row-wise) to the intercept and this linear combination would prevent the matrix inverse from being computed (as it is singular)."

— Page 95, Feature Engineering and Selection, 2019.

Regarding ORD variables, we analyse their histograms to assess their distribution, as well as their effect over z_all_i

```python
def violhists_plot(dataframe, features, yvar, rows, cols,xsize=20, ysize=20):

    fig=plt.figure(figsize=(xsize,ysize))
    for i, feature in enumerate(features):
        ax=fig.add_subplot(rows,cols,i+1)
        ax2=ax.twinx()
        sns.violinplot(ax=ax2, x=dataframe[feature], y=yvar, alpha=0.7)
        dataframe[feature].hist(bins=20,
                                ax=ax,
                                facecolor='skyblue', alpha=0.3)

        
    fig.tight_layout()  
    plt.show()
```
