# 2. Funciones de visualización  y comprovación del árbol

```python
import matplotlib.cm as cm
from matplotlib.colors import Normalize

#Definición de una escala de color para visualizar el ITE
cmap = cm.RdYlGn
norm = Normalize(vmin=-0.5, vmax=0.5)


#Conversión de RGB a HEX
def rgb_to_hex(rgb):
    for i in range(3):
        return ('#%02x%02x%02x' % (int(rgb[0]*255), int(rgb[1]*255), int(rgb[2]*255)))

    
# Reemplazamiento del código de color dentro de la descripción del árbol    
def replace_color (dot_data, values):
    inici=0
    for i in range(len(values)):
        index=dot_data[inici:].find("fillcolor=")+11+inici
        color=dot_data[index:index+7]
        inici=index+7
        old=color
        new=rgb_to_hex(cmap(norm(values[i]))[:-1])
        dot_data=dot_data.replace(old, new, 1)
    return dot_data
```

```python
## Funciones de cálculo de CATEs para un árbol determinado 

#Cálculo de CATE dado un subconjunto de datos
def cate(df):
    t=df[df.tr==1].z_all_06 
    c=df[df.tr==0].z_all_06
    
    if((len(t)==0 or len(c)==0)):   # en caso de no poderse calcular el CATE por falta de T=1 o T=0 en el subgrupo
        cate= -999
    else:
        cate= np.mean(t)-np.mean(c)
    if len(df)!=0:
        size=(len(df)-len(t))/len(df)*100
    else:
        size=0
    return [cate, size]


#Definición del árbol con los CATEs calculados dadas la estructura de un árbol y un dataset (control i tractament)
def cate_tree(dec_tree, df):
    tree_str=dec_tree.tree_
    
    #vector con el threshold de los splits (-2 el nodo no tiene hijos)
    threshold=tree_str.threshold 

    #vector con la variable usada en cada split
    numeric_features=tree_str.feature
    num_leafs=len(tree_str.feature)
    features=[]
    for i in numeric_features:
        if i>=0:
            features.append(df.columns[i])
        else:
            features.append(-2)
    
    n_split=np.sum(threshold!=-2) #número de splits
    df_vec=[] #vector con los df de los subgrupos del df que caen en cada hoja
    df_vec=[df.copy()]*num_leafs #se indica el nodo inicial, los demás se reescriben
    index_vec=[[0]] 

    for i in range(n_split):#se puebla df_vec
        for index in index_vec[i]: #para cada nivel del árbol, se crea un vector con los hijos de las hojas del nivel
            if  threshold[index]!=-2:
                left_children=tree_str.children_left[index]
                right_children=tree_str.children_right[index]
                index_vec.append( [left_children, right_children])

                father_df=df_vec[index].copy()
                df_vec[left_children]=father_df[father_df[features[index]]<=threshold[index]].copy()
                df_vec[right_children]=father_df[father_df[features[index]]>threshold[index]].copy()
    len_vec=[len(df) for df in df_vec]
    cate_vec=[cate(df)[0] for df in df_vec] #calculem los CATEs corresponendientes a los subgrupos guardados en df_vec
    size_vec=[cate(df)[1] for df in df_vec]
    
    print(len_vec)
    return cate_vec, size_vec
    
```

```python
#Cambio de los valores de las hojas de un árbol por los valores dados 
#(se usa para cambiar las medias de ITE por los valores de CATE calculados)
def replace_values (dot_data, new_values):
    inici=0
    for i in range(len(new_values)):
        index=dot_data[inici:].find("nvalue = ")+9+inici
        word=dot_data[index:index+6]
        long=len([int(s) for s in list(word) if s.isdigit()])
        signe=len([1 for s in list(word) if s=="-"])
        number=dot_data[index:index+long+1+signe]
        inici=index+long+1+signe
        old="nvalue = "+number
        new="nvalue = "+str(new_values[i])
        dot_data=dot_data.replace(old, new, 1)
    return dot_data
```

# 3. Entrenamiento del *Decision Tree*

Se entrena un *Decision Tree* sobre el *dataset*, con el ITE como *output* para estratificarlo. Para su visualización se aplica la escala de color previamente definida.

```python
from sklearn import tree
from sklearn.tree import DecisionTreeRegressor
import graphviz
import os 


#Se entrena el Decision Tree sobre los datos del dataset procesado (ite_in), con el ITE como output (ite_df.ite)
dec_tree= DecisionTreeRegressor(max_depth=3, min_samples_leaf=30).fit(ite_in, ite_df.ite)

dot_data = tree.export_graphviz(dec_tree, out_file=None, 
                                feature_names=ite_in.columns, filled=True,
                               rounded=True)

#Se modifica el color según la escala establecida 
cmap_values=[row[0,0] for row in dec_tree.tree_.value]
dot_data=replace_color(dot_data, cmap_values)

#Se dibuja el gráfico
ite_dt = graphviz.Source(dot_data, format="png", filename="ite_dt") 
ite_dt
```

# 4. Validez del *Decision Tree*

A continuación, para cada árbol entrenado, se computará el valor de CATE para los subgrupos encontramos a modo de comprovación de la validez del método de estratificación por ITE predicho.

Para realizar este cálculo se importa la estructura del árbol entrenado para poblar un *dataframe* con el subconjunto clasificado en cada hoja (mediante la función *cate_tree* previamente definida) y calcular su CATE. Finalmente, para cada árbol se calculará el error relativo de cada hoja, usando el CATE calculado como valor de referencia:

$$ER = \frac{\overline{ITE}-CATE}{CATE} \cdot 100$$

Además, se calcula también el porcentaje de individuos control dentro de cada hoja, para evaluar la representatividad del valor de CATE calculado.

## 4.A Árbol completo

```python
#Se calculan los valores observacionales del CATE para cada nodo del árbol
ite_in1=ite_in.copy()
ite_in1["z_all_06"]=cleaned_df.z_all_06
ite_in1["tr"]=cleaned_df.tr

cate_values, size=cate_tree(dec_tree, ite_in1)

#Se reemplazan los valores de ITE de cada nodo del árbol por su correspondiente CATE
dot_data1=replace_values(dot_data, cate_values)
dot_data1=replace_color(dot_data1, cate_values)
cate_dt = graphviz.Source(dot_data1, format="png", filename="cate_dt") 
```

```python
cate_dt.render(directory='assets/results')
cate_dt
```

```python
ite_leafs=[row[0,0] for row in dec_tree.tree_.value]

#Finalmente se calcula el error relativo (%) de cada nodo, con CATE como referencia

error=[ite_leafs[i]-cate_values[i] for i in range(len(cate_values))]
error_df=pd.DataFrame([error,cate_values]).transpose().copy()
error_df.columns=["Error abs", "CATE"]
error_df["Error rel (%)"]=(error_df["Error abs"]/np.abs(error_df["CATE"]))*100
error_df["Control %"]=size

error_df
```

```python
error_df_ord=error_df["Error rel (%)"].sort_values()
plt.scatter(error_df_ord.index, error_df_ord.values )

plt.xlabel("Índice de hoja")
plt.ylabel("Error relativo (%)")
plt.title("Árbol Completo")


plt.savefig('assets/results/Tree_VAL_FULL.png')

plt.show()
```
